### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)  # 100 points for a smooth curve
y = x ** 2  # Square of x

# Create the plot
plt.figure(figsize=(8, 5))  # Set figure size
plt.plot(x, y, label="y = x²", color='b', linewidth=2)  # Line plot

# Add title and labels
plt.title("Line Plot of y = x²", fontsize=14)
plt.xlabel("X Values", fontsize=12)
plt.ylabel("Y Values (Squared)", fontsize=12)

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.7)  # Dashed grid lines
plt.legend()  # Show legend

# Display the plot
plt.show()
```
